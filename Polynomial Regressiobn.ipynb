```python
%matplotlib inline
from pylab import *
import numpy as np

np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

p4 = np.poly1d(np.polyfit(x, y, 4))
```

```python
import matplotlib.pyplot as plt

xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x))

print(r2)
```

```python
# Let's try different degrees for the polynomial and see what happens with the R squared and if we could be overfitting

p5 = np.poly1d(np.polyfit(x, y, 5))
p6 = np.poly1d(np.polyfit(x, y, 6))
p7 = np.poly1d(np.polyfit(x, y, 7))

plt.scatter(x, y)
plt.plot(xp, p5(xp), c='r')
plt.title("5 degree")
```

```python
plt.scatter(x, y)
plt.plot(xp, p6(xp), c='r')
plt.title("6 degree")
```

```python
plt.scatter(x, y)
plt.plot(xp, p7(xp), c='r')
plt.title("7 degree")
```

```python
r2_5 = r2_score(y, p5(x))
r2_6 = r2_score(y, p6(x))
r2_7 = r2_score(y, p7(x))

print(r2_5)
print(r2_6)
print(r2_7)
```

```python
# The R squared increases with the degree of the polynomial. This means that our model explains better the relation between the
# variables. However, we start to overfit the model to out training set. A sing of this is when you see the scatter plot
# with the predicted curve and the observed data, if the predicted curve over reacts to outliers and tries to catch
# them all, you may well be overfitting and the model wont perform well with new data that it has not been trained with.
```
