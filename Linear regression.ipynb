```python
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

scatter(pageSpeeds, purchaseAmount)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
r_value ** 2
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitline = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitline, c='r')
plt.show()
```

```python
# Let's increase the random variation on the data and see what effect it has on the R squared

pageSpeeds_2 = np.random.normal(3.0, 5.0, 1000)
purchaseAmount_2 = 100 - (pageSpeeds + np.random.normal(0, 0.5, 1000)) * 3

scatter(pageSpeeds_2, purchaseAmount_2)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds_2, purchaseAmount_2)
```

```python
r_value ** 2     

# Increasing the variation on the data produces a far worse outcome in the model.
# The model tells us that "x" (pageSpeeds) does not explain at all the variation in "y"(purchase amount) 
# By looking at the scatter plot of this new data you can see that it is not a linear relation and a linear model
# wont do a good job
```
